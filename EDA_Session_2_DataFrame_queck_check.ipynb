**ilock-loc**
- we have a drawback with **take**
- take method will give either rows or columns at a time
- but ilock and loc will give both rows and columns at  a time


**syntax**

dataframe.iloc[rows,columns]

dataframe.iloc[start:stop:step,step:stop:step]

```python
visa_df.iloc[10:15,3:7]
```
